## Key Risk Factors

```python
correlation_matrix = df.select_dtypes(include=np.number).corr()

diabetes_correlations = correlation_matrix['Diabetes_binary'].drop('Diabetes_binary')
key_risk_factors = diabetes_correlations[abs(diabetes_correlations) > 0.1]

print("Key Risk Factors:")
print(key_risk_factors)
plt.figure(figsize=(10, 6))
sns.barplot(x=key_risk_factors.index, y=key_risk_factors.values)
plt.xticks(rotation=90)
plt.xlabel("Health Indicators")
plt.ylabel("Correlation with Diabetes")
plt.title("Key Risk Factors for Diabetes")
plt.tight_layout()
plt.show()
```

Statistical tests to validate findings.  

```python
from scipy.stats import ttest_ind

# Separate the data into two groups
group0 = df[df["Diabetes_binary"] == 0]  # Non-diabetic
group1 = df[df["Diabetes_binary"] == 1]  # Diabetic

# List of continuous features to compare
continuous_features = ['BMI', 'MentHlth', 'PhysHlth', 'Age']

# Perform t-tests for each continuous feature
print("Independent Samples t-test Results:")
print("="*50)
for feature in continuous_features:
    t_stat, p_value = ttest_ind(group0[feature], group1[feature])
    print(f"\nFeature: {feature}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Significant difference between groups (p < 0.05)")
    else:
        print("No significant difference between groups")
    print("-"*40)
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df["BMI_Category"] = df["BMI"].apply(bmi_category)
df["HealthBurdenScore"] = df["MentHlth"] + df["PhysHlth"]

def age_group(age_code):
    if age_code <= 3:
        return 'Young'
    elif age_code <= 7:
        return 'Middle-aged'
    else:
        return 'Senior'
df["Age_Group"] = df["Age"].apply(age_group)
df["MetabolicSyndrome"] = ((df["HighBP"] == 1) & (df["HighChol"] == 1)).astype(int)
df["RiskyBehavior"] = ((df["Smoker"] == 1) & (df["HvyAlcoholConsump"] == 1)).astype(int)
df.to_csv("updated_health_data.csv", index=False)
```

```python
correlation = df.corr(numeric_only=True)["Diabetes_binary"].sort_values(ascending=False)
correlation = correlation[correlation.index != "Diabetes_binary"]
top_features = correlation.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_features.values, y=top_features.index, palette="viridis")
plt.title("Top 10 Features Correlated with Diabetes Risk")
plt.xlabel("Correlation with Diabetes_binary")
plt.ylabel("Feature")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Use statistical tests to validate findings

```python
from scipy.stats import chi2_contingency
categorical_features = ["HighBP", "HighChol", "CholCheck", "Smoker",
                        "PhysActivity", "Fruits", "Veggies", "HvyAlcoholConsump",
                        "AnyHealthcare", "NoDocbcCost", "GenHlth",
                        "MetabolicSyndrome", "RiskyBehavior"]

print("Chi-Square Test Results (Categorical Features):\n")
for feature in categorical_features:
    contingency_table = pd.crosstab(df[feature], df["Diabetes_binary"])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"{feature}: p-value = {p:.4f} -> {'Significant' if p < 0.05 else 'Not Significant'}")
```

### For T-Test

```python
from scipy.stats import ttest_ind
group_0 = df[df["Diabetes_binary"] == 0]
group_1 = df[df["Diabetes_binary"] == 1]
numeric_features = ["BMI", "MentHlth", "PhysHlth", "HealthBurdenScore"]

print("\nt-test Results (Numeric Features):\n")
for feature in numeric_features:
    stat, p = ttest_ind(group_0[feature], group_1[feature], equal_var=False)
    print(f"{feature}: p-value = {p:.2f} -> {'Significant' if p < 0.05 else 'Not Significant'}")
```

```python
# Calculate mean of both groups
mean_no_diabetes = group_0[feature].mean()
mean_diabetes = group_1[feature].mean()

print(f"\nAverage {feature} for non-diabetic group: {mean_no_diabetes:.2f}")
print(f"Average {feature} for diabetic group: {mean_diabetes:.2f}")

# test
your_bmi = 30

# Compare manually
print(f"\nYour BMI: {your_bmi}")
if abs(your_bmi - mean_diabetes) < abs(your_bmi - mean_no_diabetes):
    print("⚠️ Your BMI is closer to the diabetic group average.")
else:
    print("✅ Your BMI is closer to the non-diabetic group average.")
```
